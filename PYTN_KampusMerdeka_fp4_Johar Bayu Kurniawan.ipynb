## Segmentasi Pelanggan berdasarkan perilaku penggunaan Kartu Kredit mereka

Dataset notebook ini terdiri dari perilaku penggunaan kartu kredit nasabah dengan 18 fitur perilaku. Segmentasi pelanggan dapat digunakan untuk menentukan strategi pemasaran.


**Outline :**
* Pra-pemrosesan Data
* Pengelompokan menggunakan KMeans
* Interpretasi Cluster
* Visualisasi Cluster menggunakan PCA

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity

import warnings
warnings.filterwarnings(action="ignore")

# Setiap hasil yang Anda tulis ke direktori saat ini disimpan sebagai keluaran.
```

```python
data= pd.read_csv("CC GENERAL.csv")
print(data.shape)
data.head()
```

## Preprocessing Data

#### Statistik Deskriptif Data

```python
data.describe()
```

Sepertinya data memiliki banyak outlier!

#### Berurusan dengan Nilai-Nilai yang Hilang


```python
data.isnull().sum().sort_values(ascending=False).head()
```

Memasukkan nilai yang hilang dengan rata-rata

```python
data.loc[(data['MINIMUM_PAYMENTS'].isnull()==True),'MINIMUM_PAYMENTS']=data['MINIMUM_PAYMENTS'].mean()
data.loc[(data['CREDIT_LIMIT'].isnull()==True),'CREDIT_LIMIT']=data['CREDIT_LIMIT'].mean()
```

```python
data.isnull().sum().sort_values(ascending=False).head()
```


#### Menangani Outlier

Dengan menghapus outlier, kita dapat kehilangan banyak baris karena terlalu banyak outlier dalam kumpulan data. Jadi membuat rentang untuk menangani nilai ekstrim.

```python
columns=['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
        'PAYMENTS', 'MINIMUM_PAYMENTS']

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0        
    data.loc[((data[c]>0)&(data[c]<=500)),Range]=1
    data.loc[((data[c]>500)&(data[c]<=1000)),Range]=2
    data.loc[((data[c]>1000)&(data[c]<=3000)),Range]=3
    data.loc[((data[c]>3000)&(data[c]<=5000)),Range]=4
    data.loc[((data[c]>5000)&(data[c]<=10000)),Range]=5
    data.loc[((data[c]>10000)),Range]=6
 
```

```python
columns=['BALANCE_FREQUENCY', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 
         'CASH_ADVANCE_FREQUENCY', 'PRC_FULL_PAYMENT']

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=0.1)),Range]=1
    data.loc[((data[c]>0.1)&(data[c]<=0.2)),Range]=2
    data.loc[((data[c]>0.2)&(data[c]<=0.3)),Range]=3
    data.loc[((data[c]>0.3)&(data[c]<=0.4)),Range]=4
    data.loc[((data[c]>0.4)&(data[c]<=0.5)),Range]=5
    data.loc[((data[c]>0.5)&(data[c]<=0.6)),Range]=6
    data.loc[((data[c]>0.6)&(data[c]<=0.7)),Range]=7
    data.loc[((data[c]>0.7)&(data[c]<=0.8)),Range]=8
    data.loc[((data[c]>0.8)&(data[c]<=0.9)),Range]=9
    data.loc[((data[c]>0.9)&(data[c]<=1.0)),Range]=10
    
```

```python
columns=['PURCHASES_TRX', 'CASH_ADVANCE_TRX']  

for c in columns:
    
    Range=c+'_RANGE'
    data[Range]=0
    data.loc[((data[c]>0)&(data[c]<=5)),Range]=1
    data.loc[((data[c]>5)&(data[c]<=10)),Range]=2
    data.loc[((data[c]>10)&(data[c]<=15)),Range]=3
    data.loc[((data[c]>15)&(data[c]<=20)),Range]=4
    data.loc[((data[c]>20)&(data[c]<=30)),Range]=5
    data.loc[((data[c]>30)&(data[c]<=50)),Range]=6
    data.loc[((data[c]>50)&(data[c]<=100)),Range]=7
    data.loc[((data[c]>100)),Range]=8
```

```python
data.drop(['CUST_ID', 'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES',
       'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE',
       'PURCHASES_FREQUENCY',  'ONEOFF_PURCHASES_FREQUENCY',
       'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
       'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS',
       'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT' ], axis=1, inplace=True)

X= np.asarray(data)
```

Normalisasi nilai masukan.

```python
scale = StandardScaler()
X = scale.fit_transform(X)
X.shape
```

## PEMODELAN

#### Pengelompokan menggunakan Kmeans

```python
n_clusters=30
cost=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(X)
    cost.append(kmean.inertia_)  
```

```python
plt.plot(cost, 'bx-')
```

#### Memilih 6 no cluster

```python
kmean= KMeans(6)
kmean.fit(X)
labels=kmean.labels_
```

```python
clusters=pd.concat([data, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head()
```

## Interpretasi Cluster

```python
for c in clusters:
    grid= sns.FacetGrid(clusters, col='cluster')
    grid.map(plt.hist, c)
```

* **Cluster0** Orang dengan batas kredit rata-rata hingga tinggi yang melakukan semua jenis pembelian

* **Cluster1** Grup ini memiliki lebih banyak orang dengan pembayaran jatuh tempo yang lebih sering mengambil uang muka

* **Cluster2** Pengeluar uang lebih sedikit dengan batas kredit rata-rata hingga tinggi yang membeli sebagian besar dengan mencicil

* **Cluster3** Orang dengan batas kredit tinggi yang mengambil lebih banyak uang tunai di muka

* **Cluster4** Pembelanja tinggi dengan batas kredit tinggi yang melakukan pembelian mahal

* **Cluster5** Orang yang tidak menghabiskan banyak uang dan memiliki batas kredit rata-rata hingga tinggi

(Nomor cluster berubah saat dijalankan kembali)

## Visualisasi Cluster

#### Menggunakan PCA untuk mengubah data menjadi 2 dimensi untuk visualisasi

```python
dist = 1 - cosine_similarity(X)

pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green', 
          3: 'yellow', 
          4: 'orange',  
          5:'purple'}

names = {0: 'who make all type of purchases', 
         1: 'more people with due payments', 
         2: 'who purchases mostly in installments', 
         3: 'who take more cash in advance', 
         4: 'who make expensive purchases',
         5:'who don\'t spend much money'}
  
df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend()
ax.set_title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```
